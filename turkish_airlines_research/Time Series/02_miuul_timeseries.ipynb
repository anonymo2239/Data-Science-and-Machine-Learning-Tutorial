## Mevsimsellik

![image12.png](image12.png)

Bir serinin durağan olması öngörülebilirliğini ifade etmektedir.

## Cycle (Döngü)

Mevsimsellikle karıştırılabilir fakat mevsimsellik belli bir zaman aralığındaki(gün, ay, yıl) örüntüleri barındırır. Döngü ise yapısaldır ve belli olaylar çerçevesinde(politik konuşmalar, her yıl yeni iphone un çıkıp önceki iphone un fiyatının düşmesi) gerçekleşir. Döngü, çoğunlukla içinde mevsimsellik barındırsa da barındırmadığı durumlar olabilir.

Mevsimsellik belirli gün, hafta, yıl gibi süreçlerle örtüşür.
