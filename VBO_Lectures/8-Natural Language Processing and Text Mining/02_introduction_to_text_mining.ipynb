# Metin Madenciliği ve Doğal Dil İşlemeye Giriş

Metin madenciliği, büyük miktardaki metin verisinden anlamlı bilgiler, desenler ve ilişkiler çıkarmayı amaçlayan bir veri analizi sürecidir.

Doğal dil işleme (NLP) ise bilgisayarların insan dilini anlaması, yorumlaması ve üretmesi için geliştirilen yapay zekâ yöntemlerini kapsayan bir bilim dalıdır.

```python
import pandas as pd

metin = """
A Scandal in Bohemia! 01
The Red-headed League,2
A Case, of Identity 33
The Boscombe Valley Mystery4
The Five Orange Pips1
The Man with? the Twisted Lip
The Adventure of the Blue Carbuncle
The Adventure of the Speckled Band
The Adventure of the Engineer's Thumb
The Adventure of the Noble Bachelor
The Adventure of the Beryl Coronet
The Adventure of the Copper Beeches"""

metin
```

```python
v_metin = metin.split("\n")
```

```python
v = pd.Series(v_metin)
v
```

```python
metin_vektoru = v[1:len(v)]
```

```python
metin_vektoru
```

```python
mdf = pd.DataFrame(metin_vektoru, columns=["Stories"])
```

```python
mdf
```

```python
d_mdf = mdf.copy()
```

```python
d_mdf
```

```python
list1 = [1, 2, 3]
```

```python
d_mdf = d_mdf["Stories"].apply(lambda x: " ".join(x.lower() for x in x.split())) ######### çok sert
d_mdf
```

```python
d_mdf = pd.DataFrame(d_mdf, columns=["Stories"])
```

## Noktalama İşaretlerinin Silinmesi

```python
d_mdf = d_mdf["Stories"].str.replace(r"[^\w\s]", "", regex=True) #[^\w\s] = karakter ve space olmayan tüm karakterleri seç ve "" ile yer değiştir. ######################
```

```python
d_mdf
```

## Sayıların silinmesi

```python
d_mdf = d_mdf.str.replace(r"\d", "", regex=True)
```

## Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
```

```python
d_mdf['Stories'].apply(lambda x: " ".join([st.stem(i) for i in x.split()]))
```

## Lemmatization

```python
from textblob import Word
nltk.download("wordnet")
```

```python
d_mdf = d_mdf['Stories'].apply(lambda x: " ".join([Word(i).lemmatize() for i in x.split()]))
```

Lemmatization = Bir kelimenin ek almış hallerinin eklerinden ayırarak tek bir kök altında toplanması
