```python
isimler = ["ali", "veli", "ayse"]
```

```python
for i in isimler:
    print("İsim:", i, sep="_") # print ile liste elementi arasına _ koyuyor
```

```python
print(*enumerate(isimler))
```

enumerate fonksiyonu, bir liste veya dizi içindeki her bir öğenin hem indeksini (sıra numarasını) hem de değerini döndürür.

```python
for i in enumerate(isimler):
    print(i)
```

```python
for i in enumerate("isimler"):
    print(i)
```

```python
for i in enumerate(isimler, 4):
    print(i)
```
