## Boosting Yöntemleri
Zayıf öğrenicileri bir araya getirip güçlü bir öğrenici ortaya çıkarmak fikrine dayanır. (Kearns and Valiant 1990)

## Adaptive Boosting (AdaBoost)
Zayıf sınıflandırıcıların bir araya gelerek güçlü bir sınıflandırıcı oluşturması fikrini hayata geçiren algoritmadır. (Schapires and Freund 1996-1999)  
Teorik bir yapıyı algoritmaya dönüştürebildikleri için ödüle layık görülmüşlerdir.

# Gradient Boosting Machines(GBM)

AdaBoost'un sınıflandırma ve regresyon problemlerine kolayca uyarlanabilen genelleştirilmiş versiyonudur. (Friedman 2001)  
Artıklar üzerine tek bir tahminsel model formunda olan modeller serisi kurulur.

![image.png](image16.png)

Bagging yönteminde ağaçlar bağımsızdı. Boosting yöntemlerinde ise ağaçlar birbirine bağımlıdır. İlk ağaç eğitildikten sonra, hata yapan gözlemlere daha fazla ağırlık verilir ve bir sonraki ağaç bu hataları düzeltmek üzerine eğitilir. Yani, bir ağacın çıktısı, bir sonraki ağacın nasıl eğitileceğini etkiler. Bu ardışık yapı sayesinde modeller hataları daha iyi öğrenir ancak aynı zamanda daha fazla overfitting riski taşır.

![image.png](image17.png)

GBM, yani Gradient Boosting Machines, temel olarak bir Boosting algoritmasıdır ve arka planda Gradient Descent kullanır.

Gradient Boosting'in Temel Fikri:  
* İlk model (ağaç) eğitilir ve tahmin yapılır.
* Kayıp (hata) hesaplanır.
* Bu hatayı minimize etmek için Gradient Descent kullanılır.
* Hataları öğrenmek üzere yeni bir model (ağaç) eklenir.
* Yeni model, önceki modelin hatalarını azaltmaya çalışır.

## Model

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

hit = pd.read_csv('Hitters.csv')
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df['Salary']
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.25,
                                                   random_state=42)
```

```python
gbm_model = GradientBoostingRegressor()
gbm_model.fit(X_train, y_train)
```

```python
gbm_model.get_params()
```

```python
y_pred = gbm_model.predict(X_test)
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
gbm_params = {
    'learning_rate': [0.001, 0.01, 0.1, 0.2],
    'max_depth': [3, 5, 8, 50, 100],
    'n_estimators': [200, 500, 1000, 2000], # GBM algoritması için en optimum 1000 olarak görülür.
    'subsample': [1, 0.5, 0.75],
}
```

```python
from sklearn.model_selection import GridSearchCV

gbm = GradientBoostingRegressor()
gbm_cv_model = GridSearchCV(gbm, gbm_params, cv=10, n_jobs=-1, verbose=2)
gbm_cv_model.fit(X_train, y_train)
```

```python
gbm_cv_model.best_params_
```

```python
gbm_tuned = GradientBoostingRegressor(learning_rate=0.1,
                                     max_depth=3,
                                     n_estimators=2000,
                                     subsample=0.75)
gbm_tuned.fit(X_train, y_train)
```

```python
y_pred = gbm_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
import matplotlib.pyplot as plt
Importance = pd.DataFrame({'Importance': gbm_tuned.feature_importances_*100},
                         index=X_train.columns)
Importance.sort_values(by='Importance', axis=0, ascending=True).plot(kind='barh', color='r')
plt.xlabel('Değişkenlerin Önem Düzeyleri')
```
