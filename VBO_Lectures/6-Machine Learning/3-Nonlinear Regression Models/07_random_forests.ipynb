## Model

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

hit = pd.read_csv('Hitters.csv')
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df['Salary']
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.25,
                                                   random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state=23)
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_model.get_params()
```

En önemli parametreler, fit edilecek ağaç sayısı ve max_featuresdir.

1. **`max_depth`: None**
   - Maksimum derinliktir. 
   - `None` olarak ayarlandığında, ağaç tam olarak genişleyene kadar büyür. 
   - Aşırı öğrenmeyi önlemek için genellikle sınırlandırılır.

---

2. **`max_features`: 1.0**
   - Her düğümde dallanma yaparken kullanılacak maksimum değişken sayısını belirtir.
   - `1.0` olarak ayarlanmışsa, tüm özellikler kullanılacaktır.
   - Daha küçük bir değer (örneğin, `sqrt(p)` veya `log2(p)`) genellikle daha fazla çeşitlilik sağlar.

---

3. **`max_leaf_nodes`: None**
   - Maksimum yaprak düğümü sayısını belirler.
   - `None` olduğunda sınırsız yaprak düğümü oluşturur.
   - Aşırı büyümeyi engellemek için bu değeri sınırlamak faydalı olabilir.

---

4. **`max_samples`: None**
   - Her bir ağacı oluşturmak için kullanılacak örnek sayısını belirler.
   - `None` olduğunda, tüm veri kümesi kullanılır (bootstrap örnekleme ile).
   - Daha küçük bir oran belirleyerek daha hızlı eğitim sağlanabilir.

---

5. **`min_samples_leaf`: 1**
   - Bir yaprak düğümünde bulunması gereken minimum örnek sayısını belirtir.
   - `1` olarak ayarlandığında, yaprak düğümünde tek bir örnek bile bulunabilir.
   - Daha yüksek değerler aşırı öğrenmeyi önler.

---

6. **`min_samples_split`: 2**
   - Bir düğümün dallanabilmesi için içerdiği minimum örnek sayısını belirtir.
   - `2` olarak ayarlanmışsa, en az iki örnek olduğunda dallanma yapılır.

---

7. **`min_weight_fraction_leaf`: 0.0**
   - Yaprak düğümünde bulunması gereken minimum ağırlık oranını belirtir.
   - Ağırlıklı veri kümeleri kullanıldığında önemlidir.
   - `0.0` olarak ayarlandığında herhangi bir minimum ağırlık şartı yoktur.

---

8. **`n_estimators`: 100**
   - Ormanda oluşturulacak karar ağacı sayısını belirtir.
   - Daha yüksek değerler daha sağlam ve genelleme yeteneği yüksek modeller oluşturur, ancak eğitim süresini artırır.

---

## Tahmin

```python
rf_model.predict(X_test)[0:5]
```

```python
y_pred = rf_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```

## Model Tuning

```python
from sklearn.model_selection import GridSearchCV
rf_params = {'max_depth': list(range(1, 20)),
            'max_features': [3, 5, 10, 15],
            'n_estimators': [200, 500, 1000, 2000]}  # Random Forests'ta kabul gören değerler bunlar
```

```python
rf_model = RandomForestRegressor(random_state=43)
```

```python
rf_cv_model = GridSearchCV(rf_model,
                           rf_params,
                           cv=10, 
                           n_jobs=-1)
# 3 özelliğin parametrelerini girdik. Bunların içindeki değerlerin fazlalığı, bu özelliklerin kombinasyonları
# ve ağaç derinliğini hesaba katacak olursak best_params bulma çok uzun süren bir işlem olacaktır.
# n_jobs=-1 parametresi, işlemcilerin tam kapasite kullanılmasını sağlıyor.
```

```python
rf_cv_model.fit(X_train, y_train)
```

```python
rf_tuned = RandomForestRegressor(max_depth=4,
                                max_features=3,
                                n_estimators=2000)
```

```python
rf_tuned.fit(X_train, y_train)
```

```python
y_pred = rf_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_pred, y_test))
```

Random Forests ile Bagging birbirine yakın çıktı çünkü ağaçlar bagging yöntemindeki gibi bootstrap ile seçilir.

Random Forests, Bagging'in özel bir uygulamasıdır ve sadece karar ağaçlarını kullanır, ayrıca değişken rastgeleleştirme ile daha çeşitli modeller oluşturur.
