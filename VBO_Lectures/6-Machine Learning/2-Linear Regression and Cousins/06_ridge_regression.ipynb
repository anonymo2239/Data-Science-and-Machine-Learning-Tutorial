# Ridge Regression

Amaç; hata kareler toplamını minimize eden katsayıları, bu katsayılara bir ceza uygulayarak bulmaktır.

![image.png](imagee22.png)

Hata kareler toplamı

EKK (En Küçük Kareler - Ordinary Least Squares, OLS): Regresyon katsayılarını belirlemek için kullanılan yöntemdir. Amaç, tahmin edilen değerler ile gerçek değerler arasındaki farkların kareler toplamını minimize etmektir.

SSE (Hata Kareler Toplamı - Sum of Squared Errors): Modelin hatasını ölçen bir metriktir. Tahmin edilen değerler ile gerçek değerler arasındaki farkların kareleri toplamıdır.

Sonuç: OLS, regresyon katsayılarını belirlemek için kullanılan bir yöntemdir ve SSE’yi minimize etmeye çalışır, ancak ikisi aynı şey değildir.

![image.png](imagee23.png)

L2 düzenleştirme yöntemidir. Bu yöntem, modelin tüm katsayılarını küçültür, ancak hiçbiri tam olarak sıfıra inmez. Bu, modelin overfitting riskini azaltırken, tüm özelliklerin katkısını korur.

Bazı kaynaklarda- Ridge, Lasso, ElasticNet yöntemleri- Düzenlileştirme Yöntemleri olarak da geçer.

Lambda parametresi, uygulanacak cezanın şiddetini ayarlamaya yarar. Belli yöntemler ile bu parametreyi belirleyeceğiz. Bazı kaynaklarda Tuning parametresi, daraltma faktörü vs. olarak geçebilir.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np 

hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float')
X = pd.concat([X_, ms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=0.1).fit(X_train, y_train) # Buradaki alpha, lambda katsayısı
```

```python
ridge_model.coef_
```

```python
from matplotlib.pyplot import plot
import matplotlib.pyplot as plt

lambdalar = 10**np.linspace(10,-2,100)*0.5

ridge_model = Ridge()
katsayilar = []

for i in lambdalar:
    ridge_model.set_params(alpha = i)
    ridge_model.fit(X_train, y_train)
    katsayilar.append(ridge_model.coef_)

ax = plt.gca()
ax.plot(lambdalar, katsayilar)
ax.set_xscale('log')

plt.xlabel('Lambda(Alpha) Değerleri')
plt.ylabel('Katsayılar/Ağırlıklar')
plt.title('Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları');
```

Ridge regresyonu katsayıları sıfır yapmaz. Katsayıları sıfıra yaklaştırır ama hiçbir zaman modelden çıkarmaz ve Lasso'dan farkı budur. Katsayısı büyük olmaması gereken bir değişkene büyük bir ceza uygular ve o değişkenin katsayısını sıfıra yaklaştırır. Bu özellik Ridge regresyonunun temel amacıdır - model karmaşıklığını azaltmak ve daha genellenebilir bir model elde etmek.

## Tahmin

```python
y_pred = ridge_model.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
r2_score(y_test, y_pred)
```

Bu da ortalama bir tahminden daha iyi bir performans sergilediğini gösterir. Ancak, bu sonuç yine de oldukça düşük; genellikle daha yüksek bir R² değeri istenir.

## Model Tuning

```python
lambdalar = 10**np.linspace(10, -2, 100)*0.5
lambdalar[0:5]
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

scaler = StandardScaler()
ridge_cv = RidgeCV(alphas=lambdalar, scoring="neg_mean_squared_error") # optimum alpha değerini cross val ile bulmaya yarıyor
pipeline = make_pipeline(scaler, ridge_cv)
pipeline.fit(X_train, y_train)
```

```python
ridge_cv.alpha_
```

```python
scaler = StandardScaler()
ridge_tuned = Ridge(alpha=ridge_cv.alpha_)
pipeline = make_pipeline(scaler, ridge_cv)
pipeline.fit(X_train, y_train)
```

```python
np.sqrt(mean_squared_error(y_test, pipeline.predict(X_test)))
```
