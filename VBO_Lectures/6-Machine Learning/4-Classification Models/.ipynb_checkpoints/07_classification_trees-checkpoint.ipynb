# Classification and Regression Trees (CART)

Amaç, veri seti içerisindeki karmaşık yapıları basit karar yapılarına dönüştürmektir. Heterojen veri setleri belirlenmiş bir hedef değişkene göre homojen alt gruplara ayrılır.

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import GridSearchCV

diabetes = pd.read_csv('diabetes.csv')
df = diabetes.copy()
df = df.dropna()
y = df['Outcome']
X = df.drop('Outcome', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3,
                                                   random_state=238)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cart = DecisionTreeClassifier()
cart_model = cart.fit(X_train, y_train)
```

```python
y_pred = cart_model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

İki parametrenin farkı şudur:

* **`min_samples_split`**: Bir iç düğümün **bölünebilmesi için** gereken minimum örnek sayısıdır.
  Örn: `min_samples_split=4` ise, bir düğümde 4'ten az örnek varsa bölünmez.

* **`min_samples_leaf`**: Bir **yaprak düğümde kalması gereken** minimum örnek sayısıdır.
  Örn: `min_samples_leaf=2` ise, herhangi bir dal sonunda en az 2 örnek olmalı.

### Kısaca:

* `split`: **bölünme koşulu**
* `leaf`: **yaprakta kalma şartı**

**`max_depth`**: Karar ağacının alabileceği **maksimum derinliği** belirler.

* Derinlik = kökten yaprağa kadar olan maksimum seviye sayısı.
* `max_depth` sınırı konmazsa ağaç çok derinleşip **overfitting** yapabilir.
* Küçük bir `max_depth` daha basit ama genellemesi iyi bir model sağlar.

### Özet:

* `max_depth`: ağacın maksimum katman sayısı.
* Aşırı büyük → overfit,
* Aşırı küçük → underfit riski.

```python
cart_grid = {'max_depth': list(range(1, 20)), 
             'min_samples_split': list(range(2, 50)),
             'min_samples_leaf': list(range(1, 20))}
```

```python
cart = DecisionTreeClassifier()
cart_cv_model = GridSearchCV(cart, cart_grid, cv=10, n_jobs=-1, verbose=2)
cart_cv_model.fit(X_train, y_train)
```

```python
cart_cv_model.best_estimator_
```

```python
cart_cv_model.best_estimator_.min_samples_split
```

```python
cart_tuned = DecisionTreeClassifier(max_depth=5, min_samples_leaf=17, min_samples_split=2)
```

```python
cart_tuned.fit(X_train, y_train)
```

```python
y_train_pred = cart_tuned.predict(X_train)
acc_train = accuracy_score(y_train, y_train_pred)
acc_train
```

```python
y_test_pred = cart_tuned.predict(X_test) 
acc_test = accuracy_score(y_test, y_test_pred)
acc_test
```
