# Logistic Regression

Amaç, sınıflandırma problemi için bağımlı ve bağımsız değişkenler arasındaki ilişkiyi tanımlayan doğrusal bir model kurmaktır.

### Neden sınıflandırmada lineer regresyon değil de lojistik regresyon kullanıyoruz?

Lojistik regresyon ile lineer regresyon arasındaki temel farkı basitçe açıklayayım:

1. Lineer regresyon sürekli değerler (örn: ev fiyatı, maaş) tahmin ederken, lojistik regresyon 0-1 arasında olasılık değerleri üretir (örn: müşteri ayrılır mı/ayrılmaz mı).

2. Lineer regresyon formülü:
   - y = b0 + b1x  (sınırsız değerler üretebilir)

3. Lojistik regresyon formülü:
   - p = 1 / (1 + e^(-y))  (0-1 arasında değerler üretir)

4. Neden lojistik tercih edilir:
   - Olasılık değerleri üretir (0-1 arası)
   - Sınıflandırma için daha uygundur
   - Sonuçları yorumlamak daha kolaydır
   - S şeklinde bir eğri (sigmoid) kullanır ki bu ikili sınıflandırma için idealdir

Özetle, lojistik regresyon sınıflandırma problemleri için özel olarak tasarlanmış bir yöntemdir ve bu tür problemlerde lineer regresyondan daha iyi sonuç verir.

Lineer regresyon bazı sınıflandırma problemlerinde kullanılabilir, ama önemli kısıtlamaları var:

KULLANILDIĞI DURUMLAR:
* Basit ikili sınıflandırma problemlerinde
* Veri çok net bir şekilde ayrılabiliyorsa

**Lojistik regresyon, regresyon problemleri için kullanılmaz.**

## Model

````python
import pandas as pd 
import numpy as np 

diabetes = pd.read_csv("diabetes.csv")
df = diabetes.copy()
df = df.dropna()
df.head()
````

````python
df.info()
````

````python
df["Outcome"].value_counts()
````

````python
df["Outcome"].value_counts().plot.barh();
````

````python
df.describe().T
````

````python
y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
````

````python
!pip install statsmodels
````

### statsmodels

````python
import statsmodels.api as sm

log = sm.Logit(y, X)
log_model = log.fit()
log_model.summary()
````

### scikit-learn

````python
from sklearn.linear_model import LogisticRegression 
log = LogisticRegression(solver="liblinear")
log_model = log.fit(X, y)
log_model
````

````python
log_model.intercept_
````

````python
log_model.coef_
````

## Tahmin & Model Tuning

Sınıflandırma problemlerinde bağımlı değişkenlerde ilgilendiğimiz sınıfın doğru sınıf olup olmama durumunu kontrol etmeliyiz. Örneğin diyabet riski 'olması' durumu, kanser 'olması' durumu veya kredisini 'ödeyememe' durumuna odaklanılır, yani olasılığı daha düşük durumlara odaklanılır ve bu durumlara genelde 1 durumu verilir. Bu durumu kontrol etmessek ve yanlış anlarsak tüm projede yanlış devam etmiş oluruz.

````python
y_pred = log_model.predict(X)
````

````python
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix, classification_report

confusion_matrix(y, y_pred)
````

```
            Tahmin 0    Tahmin 1
Gerçek 0     448         52
Gerçek 1     121         147
```

- **448**: Gerçekten 0 olanları doğru 0 tahmin ettik (**True Negative**).  
- **52**: Gerçekten 0 olanları yanlışlıkla 1 tahmin ettik (**False Positive**).  
- **121**: Gerçekten 1 olanları yanlışlıkla 0 tahmin ettik (**False Negative**).  
- **147**: Gerçekten 1 olanları doğru 1 tahmin ettik (**True Positive**).

---

Özet:
- **True Positive (TP)** = 147
- **True Negative (TN)** = 448
- **False Positive (FP)** = 52
- **False Negative (FN)** = 121

Logistic Regression'da tune edebileceğimiz bir durum yok fakat cross validation ile doğrulayabiliriz.

````python
accuracy_score(y, y_pred)
````

````python
print(classification_report(y, y_pred))
````

````python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

logit_roca_auc = roc_auc_score(y, log_model.predict(X))
fpr, tpr, thresholds = roc_curve(y, log_model.predict_proba(X)[:,1])

plt.figure()
plt.plot(fpr, tpr, label='AUC (area = %0.2f)' % logit_roca_auc)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Oranı')
plt.ylabel('True Positive Oranı')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
````

Bu kod bir **ROC Eğrisi (Receiver Operating Characteristic Curve)** çizer.

**Adım adım:**
- `roc_auc_score` ile modelin AUC (Area Under Curve) skorunu hesaplıyor. Yani modelin sınıflandırmadaki genel başarısını özetliyor. (1'e ne kadar yakınsa o kadar iyi)
- `roc_curve` ile farklı eşik değerleri için **True Positive Rate (TPR)** ve **False Positive Rate (FPR)** değerlerini çıkarıyor.
- `[0,1]`'den `[0,1]`'e çizilen kırmızı kesikli çizgi (**r--**) "rastgele tahmin" sınırını temsil ediyor.

---

**Özetle:**  
Bu grafik, bir modelin sınıflandırmadaki başarı ve başarısızlık oranlarını görsel olarak anlamamızı sağlar.  
Eğri ne kadar sol üst köşeye yakınsa model o kadar iyi çalışıyor demektir.

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=21) 
````

````python
log2 = LogisticRegression(solver='liblinear')
log_model2 = log2.fit(X_train, y_train)
log_model2
````

````python
accuracy_score(y_test, log_model2.predict(X_test))
````

````python
from sklearn.model_selection import cross_val_score

cross_val_score(log_model2, X_test, y_test, cv=10).mean()
````

Bu kod:

> **Modelini (`log_model2`) 10 farklı parçaya (fold) bölerek test ediyor ve bu 10 testin ortalama skorunu alıyor.**

---

### Adım adım:

- `cross_val_score(...)` ➔ 10 farklı **cross-validation** testi yapıyor (cv=10 demek bu).
- Her seferinde farklı bir 9 parça ile eğitip, kalan 1 parça ile test ediyor.
- 10 tane sonuç çıkıyor.
- `.mean()` ➔ bu 10 sonucun ortalamasını alıyor.

---

### Sonuç olarak:
✅ Modelin daha güvenilir ve genel geçer bir performansını ölçüyorsun.  
✅ Tek bir bölmeye bağlı kalmadan **ortalama başarı** görüyorsun.
