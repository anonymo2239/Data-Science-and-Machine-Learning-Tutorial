## Model

```python
import pandas as pd 
from sklearn.model_selection import train_test_split

diabetes = pd.read_csv('diabetes.csv')
df = diabetes.copy()
df = df.dropna()
y = df['Outcome']
X = df.drop('Outcome', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3,
                                                   random_state=238)
```

```python
from sklearn.naive_bayes import GaussianNB
```

Naive Bayes algoritmasında da tune edilebilecek herhangi bir hiperparametre yok. O yüzden doğrulayıp devam edeceğiz.

```python
nb_model = GaussianNB().fit(X_train, y_train)
nb_model
```

```python
nb_model.predict(X_test)[0:10]
```

**NOT:** Train-test ayrımı yaparken sınıflar arası dağılıma önem vermeliyiz. Şayet 0 sınıfında 1000, 1 sınıfında 10000 adet örnek varsa bu dağılım çok dengesiz olur ve 0 sınıfını yeteri kadar göremeyebileceği ihtimaline karşın test oranını artırabiliriz. Örneğin test_size parametresini 0.25 yerine 0.30 0.40 vs. yapabiliriz.

```python
nb_model.predict_proba(X_test)[0:10]
```

```python
y_pred = nb_model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
cross_val_score(nb_model, X_test, y_test, cv=10).mean()
```

Eğer verin küçükse → cv=5 veya cv=10 seçmek en iyisidir.  
Eğer verin büyükse → cv=5 çoğunlukla yeterlidir, daha fazlası aşırıya kaçar.  
Çok büyük veri setlerinde → cv = 3 bile bazen yeterli olur, çünkü zaten veri çok fazla.

Burada hiperparametre tuning den ziyade doğruluk oranını 'doğrulamış' olduk.
