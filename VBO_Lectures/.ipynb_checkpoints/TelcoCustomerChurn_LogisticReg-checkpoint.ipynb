```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/Churn.csv")
```

```python
df.head()
```

```python
df.T
```

```python
df.shape
```

```python
# Veri temizleme iÅŸlemleri
```

```python
df.dtypes # TotalCharges'e object demiÅŸler. Bunun sebebi bazÄ± deÄŸerlerde kesme iÅŸareti vb. olabilir.
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors="coerce") #errors="coerce" hatalÄ± olanlarÄ± NaN yapar. NaN Ä±n aÃ§Ä±lÄ±mÄ± Not a Number'dÄ±r.
```

```python
df.isnull().sum() # TotalCharges'da 11 tane NaN var. BunnlarÄ± dolduralÄ±m.
```

```python
df.TotalCharges = df.TotalCharges.fillna(df.TotalCharges.mean())
```

```python
df.isnull().sum().sum() # NaN kalmadÄ±.
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
# Bu satÄ±rda Ã¶nce bÃ¼tÃ¼n sÃ¼tun isimlerini kÃ¼Ã§Ã¼k harfe Ã§evirip sonra boÅŸluklarÄ± _ ile deÄŸiÅŸtirdik.
```

```python
string_columns = list(df.dtypes[df.dtypes=="object"].index) # Obje tipindeki dosyalarÄ± seÃ§tik
```

```python
string_columns
```

```python
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(" ", "_")

# BÃ¶ylelikle bÃ¼tÃ¼n string sÃ¼tunlar kÃ¼Ã§Ã¼k harfe Ã§evrildi ve boÅŸluklar _ ile deÄŸiÅŸtirildi.
```

```python
df.head()
```

```python
df.churn.head()
```

```python
df.churn = (df.churn == "yes").astype(int) # Churn sÃ¼tunundaki deÄŸerleri evet ise 1, hayÄ±r ise 0 yapar.
```

```python
df.churn.head()
```

```python
# VERÄ° SETÄ°NÄ° PARÃ‡ALAMA
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train_full, df_test = train_test_split(df, test_size=20, random_state=42) # random_state=42 ile her seferinde aynÄ± ÅŸekilde bÃ¶lÃ¼nmesini saÄŸladÄ±k.
```

```python
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)
```

```python
# HEDEF DEÄžÄ°ÅžKEN VE Ã–ZNÄ°TELÄ°K BELÄ°RLEME
```

```python
y_train = df_train.churn.values
```

```python
y_val = df_val.churn.values
```

```python
del df_train["churn"]
del df_val["churn"]
# Bu satÄ±r ile train ve val setlerinden churn sÃ¼tununu sildik.
```

```python
# FEATURE ENGINEERING 
# Model kurmadan Ã¶nce veri setimizi dÃ¼zelttik. Åžimdi one hot encoding yapacaÄŸÄ±z.
```

```python
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents',
               'phoneservice', 'multiplelines', 'internetservice',
               'onlinesecurity', 'onlinebackup', 'deviceprotection',
               'techsupport', 'streamingtv', 'streamingmovies', 'contract',
               'paperlessbilling', 'paymentmethod']
```

```python
numerical = ['tenure', 'monthlycharges', 'totalcharges']
# Bu satÄ±rda kategorik ve sayÄ±sal sÃ¼tunlarÄ± ayÄ±rdÄ±k. Ã‡Ã¼nkÃ¼ kategorik sÃ¼tunlarÄ± one hot encoding yapacaÄŸÄ±z.
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient="records")
# Bu satÄ±rda kategorik ve sayÄ±sal sÃ¼tunlarÄ± birleÅŸtirdik. Buradaki orient="records" satÄ±rÄ±  her bir satÄ±rÄ± bir sÃ¶zlÃ¼k yapar.
```

```python
train_dict[:1]
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)
```

```python
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train[0]
```

```python
dv.get_feature_names_out()
```

```python
# ArtÄ±k Ã¶znitelik mÃ¼hendisliÄŸi iÅŸlemlerini tamamladÄ±k. Ã‡iÃ§ek gibi bir veri setimiz var. Åžimdi model kurmaya geÃ§ebiliriz.
```

```python
# MODEL KURMA
```

```python
# EÄŸer iyi bir veri setiniz varsa iyi bir model kurarsÄ±nÄ±z. Buna yazÄ±lÄ±m biliminde garbage in garbage out denir. Ã‡Ã¼nkÃ¼ kÃ¶tÃ¼ bir veri seti ile iyi bir model kurulamaz.
```

```python
# BilindiÄŸi Ã¼zere regresyon modelleri hedef deÄŸiÅŸkenin sayÄ±sal olduÄŸu durumlarda kullanÄ±lÄ±r. 
# Logistic resgression bir lineer modeldir ama sÄ±nÄ±fsal problemler iÃ§in kullanÄ±lÄ±r. (Sigmoit fonksiyonu kullanÄ±r.)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver="liblinear", random_state=42) # Bu sÄ±nÄ±fdan Ã¶rnek aldÄ±k. Solver parametresi optimizasyon iÃ§in kullanÄ±lÄ±r. KÃ¼Ã§Ã¼k Ã¶rneklemler iÃ§in liblinear kullanÄ±lmasÄ± tavsiye edilir.
```

```python
model.fit(X_train, y_train)
```

```python
# MODEL DEÄžERLENDÄ°RME
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient="records")
```

```python
X_val = dv.transform(val_dict)
# Bu satÄ±rda val setini one hot encoding yaparak modelimizi deÄŸerlendireceÄŸiz.
```

```python
y_pred = model.predict_proba(X_val)
# predict_proba fonksiyonu ile modelin tahmin ettiÄŸi olasÄ±lÄ±klarÄ± alÄ±rÄ±z.
```

```python
y_pred[:5]
# Bu satÄ±rda ilk 5 deÄŸeri aldÄ±k. Ä°lk sÃ¼tun 0 olma olasÄ±lÄ±ÄŸÄ±, ikinci sÃ¼tun 1 olma olasÄ±lÄ±ÄŸÄ±dÄ±r.
# Ã–rneÄŸin ilk satÄ±rda mÃ¼ÅŸterinin aboneliÄŸi iptal etmeme olasÄ±lÄ±ÄŸÄ±(not churn) 0.87424, iptal etme olasÄ±lÄ±ÄŸÄ± 0.1257'tir.
```

```python
model.score(X_val, y_val)
```

```python
model.score(X_train, y_train)
```

```python
# Skor ne kadar birbirine yakÄ±nsa ve her ikisi bire yakÄ±nsa model o kadar iyidir.
```
