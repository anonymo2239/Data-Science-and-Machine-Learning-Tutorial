```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/Churn.csv")
```

```python
df.head()
```

```python
df.T
```

```python
df.shape
```

```python
# Veri temizleme işlemleri
```

```python
df.dtypes # TotalCharges'e object demişler. Bunun sebebi bazı değerlerde kesme işareti vb. olabilir.
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors="coerce") #errors="coerce" hatalı olanları NaN yapar. NaN ın açılımı Not a Number'dır.
```

```python
df.isnull().sum() # TotalCharges'da 11 tane NaN var. Bunnları dolduralım.
```

```python
df.TotalCharges = df.TotalCharges.fillna(df.TotalCharges.mean())
```

```python
df.isnull().sum().sum() # NaN kalmadı.
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
# Bu satırda önce bütün sütun isimlerini küçük harfe çevirip sonra boşlukları _ ile değiştirdik.
```

```python
string_columns = list(df.dtypes[df.dtypes=="object"].index) # Obje tipindeki dosyaları seçtik
```

```python
string_columns
```

```python
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(" ", "_")

# Böylelikle bütün string sütunlar küçük harfe çevrildi ve boşluklar _ ile değiştirildi.
```

```python
df.head()
```

```python
df.churn.head()
```

```python
df.churn = (df.churn == "yes").astype(int) # Churn sütunundaki değerleri evet ise 1, hayır ise 0 yapar.
```

```python
df.churn.head()
```

```python
# VERİ SETİNİ PARÇALAMA
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train_full, df_test = train_test_split(df, test_size=20, random_state=42) # random_state=42 ile her seferinde aynı şekilde bölünmesini sağladık.
```

```python
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)
```

```python
# HEDEF DEĞİŞKEN VE ÖZNİTELİK BELİRLEME
```

```python
y_train = df_train.churn.values
```

```python
y_val = df_val.churn.values
```

```python
del df_train["churn"]
del df_val["churn"]
# Bu satır ile train ve val setlerinden churn sütununu sildik.
```

```python
# FEATURE ENGINEERING 
# Model kurmadan önce veri setimizi düzelttik. Şimdi one hot encoding yapacağız.
```

```python
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents',
               'phoneservice', 'multiplelines', 'internetservice',
               'onlinesecurity', 'onlinebackup', 'deviceprotection',
               'techsupport', 'streamingtv', 'streamingmovies', 'contract',
               'paperlessbilling', 'paymentmethod']
```

```python
numerical = ['tenure', 'monthlycharges', 'totalcharges']
# Bu satırda kategorik ve sayısal sütunları ayırdık. Çünkü kategorik sütunları one hot encoding yapacağız.
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient="records")
# Bu satırda kategorik ve sayısal sütunları birleştirdik. Buradaki orient="records" satırı  her bir satırı bir sözlük yapar.
```

```python
train_dict[:1]
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)
```

```python
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train[0]
```

```python
dv.get_feature_names_out()
```

```python
# Artık öznitelik mühendisliği işlemlerini tamamladık. Çiçek gibi bir veri setimiz var. Şimdi model kurmaya geçebiliriz.
```

```python
# MODEL KURMA
```

```python
# Eğer iyi bir veri setiniz varsa iyi bir model kurarsınız. Buna yazılım biliminde garbage in garbage out denir. Çünkü kötü bir veri seti ile iyi bir model kurulamaz.
```

```python
# Bilindiği üzere regresyon modelleri hedef değişkenin sayısal olduğu durumlarda kullanılır. 
# Logistic resgression bir lineer modeldir ama sınıfsal problemler için kullanılır. (Sigmoit fonksiyonu kullanır.)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver="liblinear", random_state=42) # Bu sınıfdan örnek aldık. Solver parametresi optimizasyon için kullanılır. Küçük örneklemler için liblinear kullanılması tavsiye edilir.
```

```python
model.fit(X_train, y_train)
```

```python
# MODEL DEĞERLENDİRME
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient="records")
```

```python
X_val = dv.transform(val_dict)
# Bu satırda val setini one hot encoding yaparak modelimizi değerlendireceğiz.
```

```python
y_pred = model.predict_proba(X_val)
# predict_proba fonksiyonu ile modelin tahmin ettiği olasılıkları alırız.
```

```python
y_pred[:5]
# Bu satırda ilk 5 değeri aldık. İlk sütun 0 olma olasılığı, ikinci sütun 1 olma olasılığıdır.
# Örneğin ilk satırda müşterinin aboneliği iptal etmeme olasılığı(not churn) 0.87424, iptal etme olasılığı 0.1257'tir.
```

```python
model.score(X_val, y_val)
```

```python
model.score(X_train, y_train)
```

```python
# Skor ne kadar birbirine yakınsa ve her ikisi bire yakınsa model o kadar iyidir.
```

```python
model.intercept_[0]
```

```python
dict(zip(dv.get_feature_names_out(), model.coef_[0].round(3)))
```
