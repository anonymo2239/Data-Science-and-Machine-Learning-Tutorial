# Join Combining

```python
import numpy as np
import pandas as pd 

m = np.random.randint(1, 30, size=(5, 3))
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
df1
```

```python
df2 = df1 + 99
```

```python
df2
```

```python
pd.concat([df1, df2]) # Goruldugu uzere indexler duzensiz oldu.
```

```python
?pd.concat
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
df1.columns
```

```python
df2.columns = ["var1", "var2", "deg3"]
```

```python
df2
```

```python
df1
```

```python
condf =pd.concat([df1, df2], ignore_index=True)
```

```python
condf
```

```python
condf.fillna(condf.mean())
```

```python
pd.concat([df1, df2], join="inner") # ayni bilgiyi tasimayan tablolari birlestirmek cok mantikli olmayacaktir. Bunun icin inner parametresi ile kesisen degerleri eklemis olduk
```

```python
# concat fonksiyonunun join parametresinin iki turu vardir: inner ve outer. right ve left join icin 'merge' metodu kullanilir.
```

```python
pd.merge(df1, df2, how="left")
```

# Advanced Level Combining

```python
import pandas as pd
```

```python
# one to one combining
```

```python
df1 = pd.DataFrame({'calisanlar': ["Ali", "Veli", "Ayse", "Fatma"],
                   'grup': ["Accounting", "Engineering", "Engineering", "HR"]})
df1
```

```python
df2 = pd.DataFrame({'calisanlar': ["Ali", "Veli", "Ayse", "Fatma"],
                   'ilk_giris': ["2010", "2009", "2014", "2019"]})
df2
```

```python
pd.merge(df1, df2)
```

```python
df3 = pd.merge(df1, df2, on="calisanlar") # calisanlar sutununa gore birebir birlestiriyor
df3
```
