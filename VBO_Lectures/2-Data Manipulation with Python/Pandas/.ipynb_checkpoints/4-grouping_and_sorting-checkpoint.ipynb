```python
cats.groupby('Color').count()
```

```python
cats.groupby("Color")["Weight (kg)"].min()
```

```python
cats.groupby("Breed").apply(lambda df: df.Color.iloc[0])
# Burada türe göre grupladı ve her grubun ilk satırındaki rengini bize seri olarak verdi.
```

```python
cats.groupby(['Breed', 'Color'])[["Age (Years)", "Weight (kg)", "Gender"]].apply(lambda df: df.loc[df["Age (Years)"].idxmax()])
# türünün ve renginin en yaşlı kedilerini verir.
# Her breed ve color kombinasyonu için bir grup oluşturur.
```

```python
# groupby() yönteminin bir diğer önemli metodu olan agg(), bir DataFrame üzerinde aynı anda birden fazla işlev (fonksiyon) çalıştırmanıza olanak tanır. 
# Bu, veri seti üzerinde farklı istatistiksel özetler oluşturmak için çok kullanışlıdır.
```

```python
cats.groupby(["Breed"])["Weight (kg)"].agg([len, min, max])
```

```python
cats_reviewed = cats.groupby(['Breed', 'Color']).Gender.agg([len])
cats_reviewed
# Bu işlem, reviews veri setini tür ve renk bazında gruplandırarak, her grubun içinde kaç tane kedi sayısı olduğunu sayar ve sonucu iki düzeyli bir indeksle saklar.
```

```python
mi = cats_reviewed.index
type(mi)
```

```python
cats_reviewed.reset_index()
# reset_index() metodu, pandas'ta bir DataFrame'in mevcut indeksini sıfırlamak ve bunu veri sütunu olarak eklemek için kullanılır. 
# Özellikle, bir DataFrame üzerinde gruplama (groupby) işlemi yapıldığında veya bir multi-index oluşturulduğunda, 
# bu indeks yapısını düzleştirmek ve orijinal sütunlara geri döndürmek için kullanılır.
```

```python
cats_reviewed = cats_reviewed.reset_index()
cats_reviewed.sort_values(by='len', ascending=False)

# örneğin birman türünde bicolor renki kediden 9 adet varmış.
```

```python
cats_reviewed.sort_index()
```

```python
cats_reviewed.sort_values(by=['Breed', 'len'])
# tür ve renge göre gruplandırma yapıldı. bu gruplarda(multi-index) kaçar kedi(cinsiyet) olduğu belirlendi. multi-index istenmiyorsa reset_index() metodu uygulandı.
# Sonrasında, baştan türe göre ardından kedi sayısına göre tablo sıralandı. 
```

```python
# NOTE
# bu methodların hepsi aynı sonucu verir. ama bazı farklılıkları vardır.
# value_counts un seri adı farklı(count) olmakla beraber groupby methodu kullanılmaz.
# Ancak size ve count methodu nun verdiği serinin adı Color'dur. 
# size ve count methodunun arasındaki fark ise size methodu NaN değerleri de sayar. count methodu saymaz.
# SON OLARAK COUNT KULLANIRSAK BİZE DATAFRAME VERİR.
# SİZE KULLANIRSAK BİZE SERİE VERİR. 
cats_reviewed = cats.Color.value_counts()
cats_reviewed
```
