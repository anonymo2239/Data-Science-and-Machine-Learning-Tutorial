```python
reviews["Color"][0]
```

```python
# iloc, Pandas kütüphanesinde, DataFrame veya Series nesnelerinden veri seçmek için kullanılan bir indekse erişim yöntemidir. 
# iloc, verileri pozisyonlarına (indeks numaralarına) göre seçer. Bu yöntem, sıralı indeksleme (integer-location based indexing) olarak da bilinir.
reviews.iloc[0] # ilk satırı alır.
```

```python
reviews.iloc[:, 0] # satır yerine sütun almak için bu kullanılır. : = everything
```

```python
reviews.iloc[:3, 0] # bu ise ilk 3 satırdaki 0 indexli sütunu gösterir.
```

```python
# veya 1 ve 3. satır arasındaki 2 adet veriyi göstermek isteyebiliriz.
reviews.iloc[1:3, 0]
```

```python
# 1:4 gibi bir yöntem kullanmak yerine liste de yazabiliriz.
reviews.iloc[[1, 2, 3], 0]
```

```python
reviews.iloc[[1, 455, 56], 0]
# kısacası iloc'ta ilk parametrede satır indexi yazılır sonrasında sütun indexi yazılır.
```

```python
reviews.iloc[-5:] # son 5 satırı veriyor
```

```python
# loc, Pandas kütüphanesinde, DataFrame veya Series nesnelerinden etiketlerine (isimlerine) göre veri seçmek için kullanılan bir erişim yöntemidir.
# loc, etiket tabanlı indeksleme (label-based indexing) kullanarak belirli satır ve sütunları seçer.
# Bu yöntem, daha açıklayıcı ve okunabilir bir veri seçimi sağlar.
```

```python
reviews.loc[0, "Breed"]
```

```python
# iloc is conceptually simpler than loc because it ignores the dataset's indices. 
# When we use iloc we treat the dataset like a big matrix (a list of lists), one that we have to index into by position.
# loc, by contrast, uses the information in the indices to do its work.
```

```python
reviews.loc[0, ["Breed", "Age (Years)", "Weight (kg)"]]
```

```python
# indeksleme farklılığı olarak iloc klasik python indekslemesi kullanır. Yani reviews.iloc[0:500] dersem 0...449 (500 adet) indekslerini alacaktır.
# loc ise reviews.loc[0:500] dersem 0...500 (501 adet) indekslerini alacaktır.
```

```python
reviews.iloc[0:500]
```

```python
reviews.loc[0:500]
```
