## Silme

```python
import numpy as np
import pandas as pd

V1 = np.array([1, 3, 6, np.NaN, 7, 1, np.NaN, 9, 15])
V2 = np.array([7, np.NaN, 5, 8, 12, np.NaN, np.NaN, 2, 3])
V3 = np.array([np.NaN, 12, 5, 6, 14, 7, np.NaN, 2, 31])

df = pd.DataFrame(
    {"V1": V1,
     "V2": V2,
     "V3": V3}
)
df
```

```python
df.dropna()
```

```python
df.dropna(how="all") # hepsi eksik olanlari sil
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, how="all") # tum sutunu nan olan sutunlari siler
```

```python
df["sil_beni"] = np.nan
```

```python
df
```

```python
df.dropna(axis=1, how="all", inplace=True)
```

```python
df
```

Sayısal değişkenlerde atama

```python
df["V1"].fillna(df['V1'].mean())
```

```python
df.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
# ikinci yol
df.fillna(df.mean()[:]) # ezberimsi
```

```python
df.fillna(df.mean()['V1':'V2'])
```

```python
# ucuncu yol
```

```python
df.where(pd.notna(df), df.mean(), axis=1)
```

## Kategorik Değişken Kırılımında Değer Atama

```python
import numpy as np
import pandas as pd

V1 = np.array([1, 3, 6, np.NaN, 7, 1, np.NaN, 9, 15])
V2 = np.array([7, np.NaN, 5, 8, 12, np.NaN, np.NaN, 2, 3])
V3 = np.array([np.NaN, 12, 5, 6, 14, 7, np.NaN, 2, 31])
V4 = np.array(["IT", "IT", "IK", "IK", "IK", "IK", "IK", "IT", "IT"])

df = pd.DataFrame(
    {
        "maas": V1,
        "V2": V2,
        "V3": V3,
        "departman": V4
    }
)

df
```

```python
df.groupby("departman")["maas"].mean()
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean")) # transform("mean") 'her bir grubun' ortalamasını hesaplar.
```

```python
V1 = np.array([1, 3, 6, np.NaN, 7, 1, np.NaN, 9, 15])
V4 = np.array([np.nan, np.nan, "IK", "IK", "IK", "IK", "IK", "IT", "IT"], dtype=object)

df = pd.DataFrame(
    {
        "maas": V1,
        "departman": V4
    }
)

df
```

```python
# kategorik degiskenlerde en cok kullanılan eksik gozlem doldurma yontemi, en cok kullanilan degerle doldurmaktir.
```

```python
df['departman'].mode()
```

```python
df["departman"].fillna(df['departman'].mode()[0]) # goruldugu gibi bos olan yere en cok kullanilan deger ile doldurdu.
```

```python
df["departman"].fillna(method='bfill') # sonraki eleman ile doldurma (backfill)
```

```python
df['departman'].fillna(method='ffill')
```
