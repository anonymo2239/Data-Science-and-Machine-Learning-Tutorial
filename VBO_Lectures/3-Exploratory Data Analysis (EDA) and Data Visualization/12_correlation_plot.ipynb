**Korelasyon Grafikleri**

Korelasyon, değişkenler arasındaki ilişkiyi ifade eden istatistiksel bir terimdir.

**Scatterplot**

İki değişkenin arasındaki ilişkiyi bulmak için kullanılan en yaygın yaklaşım scatterplot'tur. (saçılım grafiği)
Sayısal değişkenler arasındaki ilişkiyi gösterir.

Eğer elimizdeki değişken kategorik ise barplot kullandık. Sayısal değişkenler için ise histogram, kdeplot, boxplot ve violin grafiklerini ele aldık. Bunlara eklemiş olduğumuz boyutlar çaprazlama anlamında eklemelerdi.

Scatterplot'ta ise iki sayısal değişkenin birlikte oluşturdukları yapıları değerlendircez. Tek değişkenli veri görselleştirmeden iki değişkenli veri görselleştirmeye geçicez

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
```

Toplam bill artikca tipslerde belli bir artis gozukuyor. Bu iliskiyi istatistiksel olarak korelasyon sayesinde gorebiliyoruz.

**Çaprazlamalar**

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df)
```

* Örneğin şirkette bir karar aldık ve tüm şirkete uygulayacağız ama örneğin departman kırılımını eklediğimizde her departman için farklı etki eden sonuçlar ortaya çıkabilir. Belki her departmana bu kararı uygulamamamız gerekir. Bu yüzden kırılım ve korelasyon önemlidir.
* Scatterplot diğer grafiklerden farklıdır. Çünkü istatistiksel bir terim olan korelasyonun görselleştirilmiş halidir.

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=df);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=df);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=df)
```
