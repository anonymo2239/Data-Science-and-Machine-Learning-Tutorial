```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
categoric_vars = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
cat_df = df.astype(CategoricalDtype(categories=categoric_vars, ordered=True))
df.head()
```

```python
sns.catplot(x="cut", y="price", data=df, palette="husl");
# hem barplot hem catplot bu tur analizler icin kullanilabilir
```

```python
sns.barplot(x="cut", y="price", hue="color", data=df)
# hue parametresi, verideki her bir kategori (cut) icin alt gruplara ayirma ve bu alt gruplari farkli renklerle gosterme islevi gorur
```

```python
# 3. bir boyut ekleyerek daha fazla "aciklanabilirlik" elde ettik. Veri gorsellestirmenin de amaci budur.

# ornegin ideal sutunu catplot ta 10 000 i gosteriyodu. ama burda 5000 civari gosteriyor.
# bunun sebebi, 3 degisken oldugu icin "renk ve kalite" kategorik degiskenlerinin ortalama fiyatini yazdiriyor.
# ustteki hata cubuklari ise standart sapmayi gosteriyor
```
