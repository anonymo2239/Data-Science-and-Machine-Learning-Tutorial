```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Bir önceki derste saçılım aracılığı ile sunulan bilgiyi bir de doğru üzerinden göstereceğiz.

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df) # linear model plot
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df)
```

GÖRÜLDÜĞÜ GİBİ BİR KIRILIM EKLENDİĞİNDE DOĞRUNUN EĞİMİ DEĞİŞTİ. MAKİNE ÖĞRENMESİ VE İSTATİSTİK KONULARINDA BUNLARI ELE ALACAĞIZ

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df)
```
