```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df); # bu grafik tum ikili degiskenlerin birbirleriyle olan korelasyon grafiklerini verir
```

* Scatterplot grafiğinde eğer oluşturulmuş olan grafik bir toz bulutu şeklindeyse bu, o iki değişken arasında bir ilişki olmadığı anlamına gelir.
* Bir scatterplot (dağılım grafiği) üzerinde veri noktaları grafiğin farklı bölgelerinde kümeleniyorsa, bu durum veride alt grupların (clusters) varlığına işaret eder. Bu kümelenmeler, veri setindeki doğal grupları veya segmentleri gösterir ve analizlerde bu grupların **ayrı ayrı** incelenmesi önemlidir.

Örnek:

Müşteri verilerinde gelir-harcama grafiğinde farklı bölgelerde kümelenmeler, farklı müşteri segmentlerini gösterebilir

```python
sns.pairplot(df, hue="species")
```

```python
# Goruldugu gibi turler zaten kumelenmis, birbiri icine girmemis
```

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"]);
```

```python
sns.pairplot(df, kind="reg");
```

```python
sns.pairplot(df, kind="reg", hue="species");
```

Petal_length-sepal_length grafiğinde her üç grupta da pozitif korelasyon (doğru orantı) vardır. Yani sepal_length arttıkça petal_length de artma eğilimi göstermektedir. Bu ilişki turuncu ve yeşil gruplarda daha belirgin görülse de, mavi grupta da pozitif bir korelasyon mevcuttur.
