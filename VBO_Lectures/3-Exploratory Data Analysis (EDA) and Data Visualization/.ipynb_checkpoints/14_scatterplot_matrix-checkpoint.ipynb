```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df); # bu grafik tum ikili degiskenlerin birbirleriyle olan korelasyon grafiklerini verir
```

* Scatterplot grafiinde eer oluturulmu olan grafik bir toz bulutu eklindeyse bu, o iki deiken arasnda bir iliki olmad anlamna gelir.
* Bir scatterplot (dalm grafii) zerinde veri noktalar grafiin farkl blgelerinde kmeleniyorsa, bu durum veride alt gruplarn (clusters) varlna iaret eder. Bu kmelenmeler, veri setindeki doal gruplar veya segmentleri gsterir ve analizlerde bu gruplarn **ayr ayr** incelenmesi nemlidir.

rnek:

Mteri verilerinde gelir-harcama grafiinde farkl blgelerde kmelenmeler, farkl mteri segmentlerini gsterebilir

```python
sns.pairplot(df, hue="species")
```

```python
# Goruldugu gibi turler zaten kumelenmis, birbiri icine girmemis
```

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"]);
```

```python
sns.pairplot(df, kind="reg");
```

```python
sns.pairplot(df, kind="reg", hue="species");
```

Petal_length-sepal_length grafiinde her  grupta da pozitif korelasyon (doru orant) vardr. Yani sepal_length arttka petal_length de artma eilimi gstermektedir. Bu iliki turuncu ve yeil gruplarda daha belirgin grlse de, mavi grupta da pozitif bir korelasyon mevcuttur.
