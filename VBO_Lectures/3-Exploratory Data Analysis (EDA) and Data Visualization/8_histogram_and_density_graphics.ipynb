```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price);
```

```python
help(sns.displot)
```

```python
sns.displot(df.price, bins=100) 
# burdaki bins parametresi histogram 
# grafigindeki aralik degerini ayarlamamiza yariyor.

# ve kind parametresini default olarak 'hist' aliyor
```

```python
sns.displot(df.price, kde=True) 
```

```python
sns.displot(df.price, kind="hist") # kind = kde yaparsak sadece kde yi gosterir. kind=hist yaparsak histogram gosterir.
```

```python
sns.kdeplot(df.price, fill=True) # altindaki dolguyu(maviligi) gostermek istiyorsak True yapmaliyiz
```

**Histograms and Density Graphics 2. Bölüm** 

Artık veriyi betimledik ve 'Ne oldu?' sorusuna yanıt aldık. Şimdi "Nasıl oldu?" sorusuna cevap aramalıyız.

Ama sadece bu grafiklerle doğru bilgiye ulaşılmaz. Başka parametreleri de incelemeliyiz.(Medyan-ortalama problemi gibi)

```python
(sns.FacetGrid(df,                    # FacetGrid, verileri kategorik degiskenlere gore alt gruplara ayirip, her grup icin ayri grafik cizmemizi saglar.
               hue='cut',
               height=5,
               xlim=(0, 10000))
.map(sns.kdeplot, 'price', fill=True)
.add_legend()                         # Hangi rengin hangi kategoriye ait oldugunu gosteren bir aciklama kutusu ekler
);
```

NOT: Bu grafik iki tepeli bir grafik olduğu için başka faktörler bu grafiğe etki ediyor diyebiliriz.

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df)
```

```python
# Ornegin cut'in premium sinifindaki E ve D renkleri fiyat olarak ayni bilgiyi tasiyor.
# Fakat diğer renkler bu ozelliklerden (fiyat acisindan) daha farkli bilgiler tasiyor.
```
