```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.groupby("timepoint")["signal"].count() # Her zaman noktasında esit sayida sinyal toplanmis ve bu da bu sinyallerin mekanik bir kaynaktan geldigini dogruluyor.
```

```python
df.groupby("signal").count()
```

Bu bilgiler ışığında, bu verinin arka tarafta bilgi taşıdığını ve farklı değişkenlerce gösterilebileceği anlaşılabilir.

```python
sns.lineplot(x="timepoint", y="signal", data=df)
```

Bu grafik, her "zaman noktasındaki" değerlerin (54 adet) ortalamasını ve standart sapmasını (güven aralığı) gösteriyor. 

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=df)
```

```python
sns.lineplot(x="timepoint", 
             y="signal", 
             hue="event", 
             style="event", 
             markers=True, dashes=False, data=df) 
# markers araya nokta ekliyor. dashes=False oldugunda tum cizgiler duzlesti. True oldugunda kesiklesti
```

```python
sns.lineplot(x="timepoint", 
             y="signal", 
             hue="region", 
             style="event", 
             data=df) 
```
