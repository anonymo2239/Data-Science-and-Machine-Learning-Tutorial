```python
import seaborn as sns
planets = sns.load_dataset("planets")
df = planets.copy()
df.head()
```

```python
import pandas as pd
df.method = pd.Categorical(df.method)
```

```python
kat_df = df.select_dtypes(include=["category"])
kat_df.head()
```

```python
df["method"].value_counts().plot.barh();
```

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
cat_df = df.select_dtypes(include=["category"])
```

```python
cat_df.cut.value_counts()
```

```python
# goruldugu gibi bu kategorik degiskenlerin ordinal turde oldugunu goruyoruz.
```

```python
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
cut_categories = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories=cut_categories, ordered=True))
```

```python
df.cut.head(1)
```

```python
df["cut"].value_counts().plot.barh().set_title("Cut Değişkeninin sınıf frekansları");
```

```python
sns.barplot(x=df.cut.value_counts(), y=df.cut.value_counts().index)
```

```python
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

Petal_length-sepal_length grafiğinde her üç grupta da pozitif korelasyon (doğru orantı) vardır. Yani sepal_length arttıkça petal_length de artma eğilimi göstermektedir. Bu ilişki turuncu ve yeşil gruplarda daha belirgin görülse de, mavi grupta da pozitif bir korelasyon mevcuttur.

```python
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df = df.pivot(index='month', columns='year', values='passengers') # Pivot, veriyi yeniden sekillendirmek (reshape) için kullanilir. 
# Uzun formattaki veriyi genis formata cevirir.
df
```

```python
sns.heatmap(df, annot=True, fmt="d", linewidths=.5, cbar=False) # fmt=d sayilari decimal sekilde gosteriyor. linewidth aralara beyaz cizgi koyuyor. cbar isi cubugunu(renk skalasini) kaldiriyor
```
