```python
# idxmax fonksiyonu, bir pandas DataFrame veya Series içindeki en yüksek değerin indeksini döndürür. 
# Yani, belirli bir sütundaki maksimum değerin hangi satırda olduğunu bulmak için kullanılır.
```

```python
bargain_idx = (reviews["Age (Years)"]/reviews["Weight (kg)"]).idxmax()
bargain_wine = reviews.loc[bargain_idx, ["Breed", "Color", "Gender"]]
```
