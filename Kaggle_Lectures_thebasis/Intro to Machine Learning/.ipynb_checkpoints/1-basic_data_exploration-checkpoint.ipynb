```python
# Verilen veriye ait sonuçlar, her sütun için 8 farklı değeri gösterir. Bu değerler şunlardır:

# count: Eksik olmayan değerlerin sayısını gösterir.
# mean: Değerlerin ortalamasıdır.
# std: Değerlerin ne kadar yayıldığını gösteren standart sapmadır.
# Ortalama (mean) 2.937997 odalı bir eve karşılık gelir.
# Standart sapma (0.955748) ise, oda sayısının bu ortalamadan yaklaşık olarak ne kadar sapabileceğini gösterir.
# Yani, oda sayılarının çoğu 2.937997 etrafında +/- 0.955748 aralığında olacaktır.
# Bu değerin yüksek ya da düşük olması, veri kümesindeki oda sayılarının ne kadar geniş bir yelpazeye yayıldığını gösterir. 
# Düşük bir standart sapma, değerlerin ortalamaya yakın olduğu anlamına gelirken, yüksek bir standart sapma, değerlerin ortalamadan daha uzak olduğunu gösterir.

# min: En küçük değeri temsil eder.
# 25% (1. çeyrek): Verilerin %25'inden büyük, %75'inden küçük olan değerdir.
# 50% (medyan): Ortanca değeri gösterir, yani verilerin yarısı bu değerden küçük, yarısı büyüktür.
# 75% (3. çeyrek): Verilerin %75'inden büyük, %25'inden küçük olan değerdir.
# max: En büyük değeri temsil eder.
# Eksik veriler, örneğin tek odalı bir evde ikinci yatak odası ölçülerinin toplanmaması gibi çeşitli nedenlerle ortaya çıkabilir.
```
