```python
import pandas as pd
```

```python
melbourneData = "C:/Users/Alperen Arda/OneDrive/Desktop/GitHub/Data-Science-and-Machine-Learning-Tutorial/Datasets/melb_data.csv"
data_houses = pd.read_csv(melbourneData)
```

```python
# Selecting Data for Modeling
data_houses.head()
```

```python
data_houses.columns
```

```python
# Eksik Değerlerin İşlenmesi ve dropna Kullanımı
# Veri setlerinde bazı gözlemler (örneğin, evler) için bazı değişkenlerin değeri eksik olabilir. 
# Bu durum, verilerin toplandığı süreçte bazı bilgilerin kaydedilmemiş olmasından kaynaklanabilir. 
# Pandas'ta bu tür eksik değerler genellikle NaN (Not a Number) veya NA (Not Available) olarak ifade edilir.

# Veri analizi yaparken eksik değerlerle karşılaştığınızda, bu değerlerle nasıl başa çıkacağınızı belirlemeniz gerekir. Bu noktada birkaç seçeneğiniz vardır:

# Eksik verileri göz ardı etmek.
# Eksik değerleri tahmin etmek veya doldurmak.
# Eksik verileri içeren satırları veya sütunları tamamen çıkarmak.
# Bu örnekte, en basit ve hızlı çözüm olarak, eksik veri içeren satırları veri setinden tamamen çıkarmak tercih edilmiştir.

# dropna Fonksiyonu
# Pandas'taki dropna() fonksiyonu, veri setindeki eksik değerleri içeren satırları veya sütunları kaldırmak için kullanılır.
```

```python

data_houses = data_houses.dropna(axis=0)
# axis=0 parametresi, eksik veri içeren satırları kaldırmak anlamına gelir. Yani, eksik veri içeren herhangi bir satır (örneğin, bir ev) veri setinden çıkarılır.
# Eğer axis=1 kullanılsaydı, eksik veri içeren sütunlar kaldırılırdı.
```

```python
# Veri alt kümelerini(data subset) seçmek için Pandas'ta iki yaygın yaklaşım vardır: nokta notasyonu (dot notation) ve sütun listesi ile seçme.

# 1. Nokta Notasyonu (Dot Notation)
# Bu yöntem, belirli bir sütunu seçmek için kullanılır. Nokta notasyonu, bir DataFrame'in belirli bir sütununa erişmenin basit bir yoludur.

# Örneğin:
```

```python
y = data_houses.Price
# Bu kodda, Price sütunu seçilerek y değişkenine atanır. y artık sadece bu sütunun verilerini içerir. 
# Bu yöntemi genellikle bir "hedef" (target) sütununu seçmek için kullanırız.
```

```python
# 2. Sütun Listesi ile Seçim
# Bu yöntem, birden fazla sütunu aynı anda seçmek istediğinizde kullanılır. Bir liste oluşturur ve bu listeyi DataFrame'e uygulayarak ilgili sütunları seçebilirsiniz.

# Örneğin:
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt']
X = data_houses[melbourne_features]
# Bu kodda, melbourne_features listesinde belirtilen sütunlar seçilir ve X değişkenine atanır. X, bu sütunlardan oluşan bir DataFrame'dir. 
# Bu yöntemi genellikle "özellikler" (features) seçmek için kullanırız.
# Özet
# Nokta Notasyonu: Tek bir sütunu seçmek için kullanılır. (df.column_name)
# Sütun Listesi ile Seçim: Birden fazla sütunu seçmek için kullanılır. (df[list_of_columns])
```

```python
# Selecting The Prediction Target

# We'll use the dot notation to select the column we want to predict, which is called the prediction target. 
# By convention, the prediction target is called y. So the code we need to save the house prices in the Melbourne data is
y = data_houses.Price
# yani makinenin neyi tahmin etmesini istiyorsak öngörü hedefine onu yazıyoruz. !!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```

```python
# Choosing "Features"

# Modelimize girdiler olarak kullanılan sütunlara "özellikler" (features) denir. 
# Bizim durumumuzda, bunlar ev fiyatını belirlemek için kullanılan sütunlar olacaktır. 
# Bazen hedef olarak belirlediğiniz sütun hariç tüm sütunları özellik olarak kullanırsınız. 
# Diğer zamanlarda, daha az özellik kullanmak daha avantajlı olabilir.

# For now, we'll build a model with only a few features. Later on you'll see how to iterate and compare models built with different features.
# We select multiple features by providing a list of column names inside brackets. Each item in that list should be a string (with quotes).

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt']
X = data_houses[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
# Visually checking your data with these commands is an important part of a data scientist's job. 
# You'll frequently find surprises in the dataset that deserve further inspection.
```

```python
# Building Your Model

# Scikit-learn kütüphanesini kullanarak modellerinizi oluşturacaksınız. Kodlama yaparken bu kütüphane, örnek kodlarda göreceğiniz gibi sklearn olarak yazılır.

# Model oluşturma ve kullanma adımları şunlardır:

# Tanımlama (Define): Model türü ne olacak? Bir karar ağacı mı? Başka bir model türü mü? Ayrıca, model türüne ilişkin bazı diğer parametreler de belirtilir.
# Eğitme (Fit): Sağlanan verilerden kalıpları yakalama. Bu, modellemenin kalbidir.
# Tahmin (Predict): Tam olarak kulağa geldiği gibi, tahmin yapmak.
# Değerlendirme (Evaluate): Modelin tahminlerinin ne kadar doğru olduğunu belirleme.
# Aşağıda, scikit-learn ile bir karar ağacı(decision tree) modelinin tanımlanması ve bu modelin özellikler ve hedef değişkenle eğitilmesine ilişkin bir örnek verilmiştir.
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define model
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
# random_state: random_state parametresi, bu rastgeleliklerin kontrol edilmesini sağlar. 
# Bir sayı belirleyerek, modelin eğitimi sırasında kullanılan rastgele sayıların aynı olmasını garantilersiniz. 
# Bu sayede, aynı verilerle aynı modeli birden fazla kez çalıştırdığınızda aynı sonuçları alırsınız.

# Belirlediğiniz sayı, modelin kalitesini etkilemez. Yani, random_state için kullandığınız sayı ne olursa olsun, 
# modelinizin performansı bu sayıya bağlı olarak değişmeyecektir. Önemli olan, aynı sonuçları elde edebilmek için her çalıştırmada aynı sayıyı kullanmaktır.
# Özetle, random_state kullanmak, modelin eğitimindeki rastgelelikleri kontrol eder ve deneylerinizin tekrar edilebilirliğini artırır.
```

```python
# Artık tahminlerde bulunmak için kullanabileceğimiz uygun bir modelimiz var.
# Uygulamada, halihazırda fiyatlarını bildiğimiz evler yerine, piyasaya çıkacak yeni evler için tahminlerde bulunmak isteyeceksiniz. 
# Ancak tahmin fonksiyonunun nasıl çalıştığını görmek için eğitim verilerinin ilk birkaç satırına yönelik tahminler yapacağız.
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
